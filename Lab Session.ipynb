```python
# 1. Create a parent class Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# 2. Create child classes inheriting from Person
class Student(Person):
    def __init__(self, name, age, student_id):
        # 3. Use super() to initialize inherited attributes
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):

        print(f"Student {self.name} has enrolled in {course_name}.")

    # 5. Override display_info() in one child class using super()
    def display_info(self):
        super().display_info()  # Displays name and age from parent
        print(f"Student ID: {self.student_id}")


class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        # 3. Use super() to initialize inherited attributes
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"Lecturer {self.name} is teaching {course_name}.")


# 4. Create objects and demonstrate functionality
    # Creating a Student object
    student1 = Student("Osei Emmanuel Yeboah", 19, "FOE.41.006.137.25")

    # Demonstrating the overridden display_info() method
    student1.display_info()

    # Demonstrating unique method
    student1.enroll_course("Strength of Materials")

    # Creating a Lecturer object
    lecturer1 = Lecturer("Dr. Matthew Cobbinah", 42, "UMaT-FOE-5501")

    # Demonstrating the inherited display_info() method
    lecturer1.display_info()

    # Demonstrating unique method
    lecturer1.teach_course("Object-Oriented Programming")
```
